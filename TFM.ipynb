```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

df = pd.read_excel('idealista.com.xlsx')
df.head()
```

```python
df['Calle']
```

```python
patron = r'([\w\s]+) en(?: calle de)? ([\w\s]+)(?:, ([\w\s]+))?, ([\w\s]+)'

# Extraer tipo de propiedad, calle, barrio y ciudad en nuevas columnas

df[['Tipo', 'Calle', 'Barrio', 'Ciudad']] = df['Calle'].str.extract(patron)
```

```python
df.drop(columns=['Calle', 'Inmobiliaria', 'Tiempo'], inplace=True)
df
```
